```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = "/Users/jshe/School/bios-10603/1UBQ-system/job-output/dist_beta_sheet.dat"

data = pd.read_csv(file, sep=" ", header=None)

m, b = np.polyfit(x=data.index, y=data.loc[:,0], deg=1)

linspace = np.linspace(min(data.index), max(data.index))

plt.plot(data.loc[:,0], linewidth=0.5)

plt.plot(linspace, linspace * m + b)

plt.xlabel("Frame")
plt.ylabel(r'$Distance \ (\AA)$')
plt.title(r'$Distance \ between \ 1UBQ \ \beta-Sheet \ Ends \ During \ Equilibration$')

plt.tight_layout()

plt.savefig("1ubq-bsheet-dist.png", dpi=300)
```
