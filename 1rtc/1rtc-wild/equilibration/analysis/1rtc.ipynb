```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = "/Users/jshe/School/bios-10603/1RTC/equilibration/1rtc-equilibration-volume.dat"

data = pd.read_csv(file, sep=" ", header=None)

plt.plot(data.loc[:,1], linewidth=0.5)

plt.xlabel("Frame")
plt.ylabel(r'$Volume \ (\AA^3)$')
plt.title("1RTC Volume Evolution During Equilibration")

plt.tight_layout()

plt.savefig("equilibration/1rtc-volume.png", dpi=300)
```

```python
file = "/Users/jshe/School/bios-10603/1RTC/equilibration/1rtc-equilibration-total.dat"

data = pd.read_csv(file, sep=" ", header=None)

plt.plot(data.loc[:,1], linewidth=0.25)

plt.xlabel("Frame")
plt.ylabel(r'$Total Energy \ (J)$')
plt.title("1RTC Total Energy Evolution During Equilibration")

plt.tight_layout()

plt.savefig("equilibration/1rtc-total.png", dpi=300)
```

```python
file = "/Users/jshe/School/bios-10603/1RTC/equilibration/1rtc-equilibration-rmsd.dat"

data = pd.read_csv(file, sep="   ", engine='python', header=None)

plt.plot(data.loc[:,0], data.loc[:,1], linewidth=0.5)

plt.xlabel("Frame")
plt.ylabel("RMSD")
plt.title("1RTC RMSD Evolution During Equilibration")

plt.tight_layout()

plt.savefig("equilibration/1rtc-rmsd.png", dpi=300)
```

```python
file = "/Users/jshe/School/bios-10603/1RTC/equilibration/RMSF.dat"

data = pd.read_csv(file, sep="\t")

plt.plot(data.loc[:,"Residue"], data.loc[:,"RMSF"], linewidth=1)

plt.xlabel("Residue")
plt.ylabel("RMSF")
plt.title("1RTC RMSF During Equilibration Stable Trajectory")

plt.tight_layout()

plt.savefig("equilibration/1rtc-rmsf.png", dpi=300)
```
